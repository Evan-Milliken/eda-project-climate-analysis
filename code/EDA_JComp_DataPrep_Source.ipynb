```python
import pandas as pd
from IPython.display import display
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
```

```python
ds1 = pd.read_excel("C:/Users/karen/OneDrive/VIT/Fourth Semester/EDA/EDA_J_Comp_Datasets/climate policies.xlsx")       
display(ds1.head())
print("Columns:", ds1.columns)
print("\nData Types:\n", ds1.dtypes)
print("\nMissing Values:\n", ds1.isnull().sum())

# Convert numeric columns only (ignoring text or categorical columns)
numeric_cols = ds1.select_dtypes(include=['number']).columns
ds1[numeric_cols] = ds1[numeric_cols].apply(pd.to_numeric, errors='coerce')
# Fill missing values
ds1[numeric_cols] = ds1[numeric_cols].fillna(ds1[numeric_cols].mean())  # Fill numerical NaNs with mean
ds1.fillna(ds1.mode().iloc[0], inplace=True)  # Fill categorical NaNs with mode

print("\nMissing Values:\n", ds1.isnull().sum())

# Drop duplicates
ds1.drop_duplicates(inplace=True)

display(ds1.head())

# Visualize the data before outlier removal
plt.figure(figsize=(10, 5))
sns.boxplot(data=ds1.select_dtypes(include=['float64', 'int64']))
plt.xticks(rotation=90)
plt.show()
# Shape of the dataset
print("Shape of Dataset:", ds1.shape)
print()

# Function to remove outliers and print IQR boundaries
def remove_outliers(ds1, column):
    Q1 = ds1[column].quantile(0.25)
    Q3 = ds1[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR    
    # Print IQR boundaries
    print(f"Column: {column}")
    print(f"  Q1: {Q1}, Q3: {Q3}")
    print(f"  Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    print(f"  Values below {lower_bound}: {ds1[ds1[column] < lower_bound][column].values}")
    print(f"  Values above {upper_bound}: {ds1[ds1[column] > upper_bound][column].values}\n")    
    # Return dataset without outliers
    return ds1[(ds1[column] >= lower_bound) & (ds1[column] <= upper_bound)]
# Apply the function for each numeric column
for col in ds1.select_dtypes(include=['float64', 'int64']).columns:
    ds1 = remove_outliers(ds1, col)

# Visualize the cleaned dataset
plt.figure(figsize=(10, 5))
sns.boxplot(data=ds1.select_dtypes(include=['float64', 'int64']))
plt.xticks(rotation=90)
plt.show()
# Final shape of the dataset
print("Final Shape of Dataset:", ds1.shape)

# Normalize & Standardize Data
scaler = StandardScaler()
ds1_scaled = ds1.copy()
ds1_scaled[ds1.select_dtypes(include=['float64', 'int64']).columns] = scaler.fit_transform(ds1.select_dtypes(include=['float64', 'int64']))

# Specify the columns to encode
columns_to_encode = ['Jurisdiction_standard']  # Replace with the names of your chosen columns
# Initialize the LabelEncoder
label_enc = LabelEncoder()
# Apply Label Encoding to the specified columns
for col in columns_to_encode:
    if col in ds1.columns:  # Ensure the column exists in the DataFrame
        ds1[col] = label_enc.fit_transform(ds1[col])

# Display the first few rows of the updated DataFrame
display(ds1.head())

ds1.to_csv("C:/Users/karen/OneDrive/VIT/Fourth Semester/EDA/EDA_J_Comp_Datasets/ds1_cleaned.csv", index=False)
```
